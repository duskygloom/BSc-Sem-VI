```python
df = pd.read_csv('assets/data_meal.csv')
```

```python
df
```

```python
plt.xlabel('Day')
plt.ylabel('Meal')

plt.plot(df['Day'], df['Week 1'], marker='o', linestyle='--')
plt.plot(df['Day'], df['Week 2'], marker='*', linestyle='-')
plt.plot(df['Day'], df['Week 3'], marker='.', linestyle=':')
plt.legend(['Week 1', 'Week 2', 'Week 3'])
```

```python
plt.xlabel('Day')
plt.ylabel('Meal')

plt.bar(df['Day'], df['Week 1'])
plt.bar(df['Day'], df['Week 2'])
plt.bar(df['Day'], df['Week 3'])
plt.legend(['Week 1', 'Week 2', 'Week 3'])
```
