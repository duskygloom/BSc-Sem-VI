```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('assets/personality_dataset.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data.fillna({
    'Time_spent_Alone': data['Time_spent_Alone'].mean(),
    'Stage_fear': data['Stage_fear'].mode()[0],
    'Social_event_attendance': data['Social_event_attendance'].median(),
    'Going_outside': data['Going_outside'].median(),
    'Drained_after_socializing': data['Drained_after_socializing'].mode()[0],
    'Friends_circle_size': data['Friends_circle_size'].median(),
    'Post_frequency': data['Post_frequency'].median()
}, inplace=True)
```

```python
data.info()
```

```python
data.head()
```

```python
data[['Stage_fear', 'Drained_after_socializing', 'Personality']] = data.apply({
    'Stage_fear': lambda x: 1 if x == 'Yes' else 0,
    'Drained_after_socializing': lambda x: 1 if x == 'Yes' else 0,
    'Personality': lambda x: 1 if x == 'Extrovert' else 0
})
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
data[[
    'Time_spent_Alone',
    'Social_event_attendance',
    'Going_outside',
    'Friends_circle_size',
    'Post_frequency'
]] = scaler.fit_transform(data[[
    'Time_spent_Alone',
    'Social_event_attendance',
    'Going_outside',
    'Friends_circle_size',
    'Post_frequency'
]])
```

```python
data.head()
```

```python
corr_matrix = data.corr()
```

```python
for i in data.columns:
    corr = corr_matrix[i]['Personality']
    if abs(corr) >= 0.5 and abs(corr) <= 0.8:
        print(i)
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(data.iloc[:, :7], data.iloc[:, 7])
```

```python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression()
model1.fit(xtrain, ytrain)
```

```python
from sklearn.metrics import f1_score

f1_score(ytest, [round(i) for i in model1.predict(xtest)])
```

```python
from sklearn.ensemble import RandomForestClassifier

model2 = RandomForestClassifier()
model2.fit(xtrain, ytrain)
```

```python
from sklearn.metrics import f1_score

f1_score(ytest, [round(i) for i in model2.predict(xtest)])
```
