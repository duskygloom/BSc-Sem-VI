```python
import pandas as pd
```

```python
df = pd.read_csv('assets/breast-cancer.csv', index_col='id')
```

```python
df.info()
```

```python
df.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
df
```

### **Encoding categories**

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
df['diagnosis'] = encoder.fit_transform(df['diagnosis'])
```

### **Feature scaling**

```python
normalizer = MinMaxScaler()
standardizer = StandardScaler()
```

```python
normal = normalizer.fit_transform(df[['radius_mean']])
print('Mean:', normal.mean())
print('Standard deviation:', normal.std())
```

```python
standard = standardizer.fit_transform(df[['radius_mean']])
print('Mean:', standard.mean())
print('Standard deviation:', standard.std())
```

```python
scaled_df = df.copy()

for i in scaled_df.columns:
    if i == 'diagnosis':
        continue
    scaled_df[i] = standardizer.fit_transform(scaled_df[[i]])
```

```python
scaled_df
```

### **Feature selection**

```python
corr_matrix = scaled_df.corr('pearson')['diagnosis'][1:]
```

```python
selected_df = scaled_df.copy()

for i in corr_matrix.keys():
    if abs(corr_matrix[i]) < 0.5 or abs(corr_matrix[i]) > 0.85:
        selected_df.drop(i, axis=1, inplace=True)
```

```python
selected_df
```

### **Split data**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    selected_df[selected_df.columns.drop('diagnosis')],
    selected_df['diagnosis'],
    test_size=0.2
)
```

### **Linear regression**

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
model1 = LinearRegression()
model1.fit(x_train, y_train)
```

```python
print('Slope:', model1.coef_)
print('Intercept:', model1.intercept_)
```

### **Logistic regression**

```python
model2 = LogisticRegression()
model2.fit(x_train, y_train)
```

### **Kth Nearest Neighbourhood**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model3 = KNeighborsClassifier()
model3.fit(x_train, y_train)
```

### **Decision tree**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model4 = DecisionTreeClassifier()
model4.fit(x_train, y_train)
```
